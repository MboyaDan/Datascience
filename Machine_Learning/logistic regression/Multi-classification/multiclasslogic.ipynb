```python
from sklearn.datasets import load_digits
%matplotlib inline
import matplotlib.pyplot as plt
#loading training set
digits = load_digits()
```

```python
#exploring the actual images
plt.gray() 
for i in range(5):
    plt.matshow(digits.images[i]) 
```

Create and train logistic regression model

```python

from sklearn.linear_model import LogisticRegression
#create a model object
model = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

splitting dataset into traininng and testinng sets

dividing into train & test- we don't want to overfit our model we don't want to bias it against our data

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data,digits.target, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
#train
model.fit(X_train, y_train)
```

Measure accuracy of our model

```python
model.score(X_test, y_test)
```

```python
digits.target[67]
```

checking where our model didn't do well the overall feelinf
Confusion Matrix checking where our model missed out
confusion matrix is just a nice was of visualising how well your model is doing

```python
#get the predicted values 
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
#y_test which is the truth  y predicted which is what your model predicted 
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

40 time the truth(image) was zero our model predicted it to be zero
