```python
# Matplotlib - data visualization
import matplotlib.pyplot as plt
# magic line - helps in proper visualization
%matplotlib inline
```

```python
# line chart

days = [1,2,3,4,5,6,7]
temperature = [25,30,22,35,24,20,27]
humidity = [40,50,34,39,45,41,42]

# plt.plot(x-axis,y-axis)
plt.plot(days,temperature,color='red',linestyle='dashdot',linewidth=1,marker='*',alpha=0.9)
```

```python
plt.plot(days,temperature,'--r*')
```

```python
# Axes label
plt.plot(days,temperature)
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Line graph')
```

```python
plt.plot(days,temperature,label="Temperature")
plt.plot(days,humidity,label="Humidity")
plt.legend()
plt.grid()
```

## BAR CHART

```python
company=["Google","Amazon","Microsoft","Twitter"]
revenue=[90,136,89,27]
profit=[40,20,34,12]

plt.bar(company,revenue,label="revenue")
plt.bar(company,profit,label='profit')

plt.xlabel('Company')
plt.ylabel('Value(M)')
plt.title('A sample bar graph')
plt.legend()
```

```python
plt.bar() # vertical bar graphs
plt.barh() # horizontal bar graphs
```

```python
plt.barh(company,revenue)
plt.ylabel('Company')
plt.xlabel('Revenue')

plt.title('A sample horizontal bar graph')
```

## Histograms

Are bar graphs where y-axis contains frequency of a variable

```python
blood_sugar = [113,85,90,150,149,93,88,115,135,82,77,129,80]

plt.hist(blood_sugar)
```

normal - 80-100
prediabetic - 100-125
diabetic > 125

```python
plt.hist(blood_sugar,bins=3)
```

```python
plt.hist(blood_sugar,bins=3,rwidth=0.9)
```

```python
# labelling
plt.hist(blood_sugar,bins=3,rwidth=0.9)
plt.xlabel('Blood Sugar Level')
plt.ylabel('Frequency')
plt.title('A sample Histogram')
```

```python
blood_sugar_1 = [67,98,89,120,133,150,84,69,89,79,120,112,100]
```

```python
plt.hist([blood_sugar,blood_sugar_1],bins=3,color=['green','red'],label=['men','women'])
plt.legend()
plt.grid()
```

## Pie Chart

```python
exp_labels=["HomeRent","Food","Internet bill","Car","Other utilities"]
exp_values=[1400,600,300,410,250]
```

```python
# initialize our pie chart
plt.pie(exp_values)
```

```python
plt.pie(exp_values,labels=exp_labels)
```

```python
plt.pie(exp_values,labels=exp_labels,autopct='%0.1f%%')
```

```python
plt.pie(exp_values,labels=exp_labels,autopct='%0.1f%%',explode=[0.1,0,0,0,0])
```

```python
# radius-enlarges the pie chart
plt.pie(exp_values,labels=exp_labels,autopct='%0.1f%%',explode=[0.1,0,0,0,0],radius=1.5)
```

```python
# shadow
plt.pie(exp_values,labels=exp_labels,autopct='%0.1f%%',explode=[0.1,0,0,0,0],radius=3,shadow=True)
```

## Data visualization from file

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('companies.csv')
```

```python
df
```

```python
col_company = df['Company']
col_Revenue = df['Revenue']
col_Profit = df['Profit']
```

```python
plt.bar(col_company,col_Revenue,label='Revenue')
plt.bar(col_company,col_Profit,label='Profit')
plt.legend()
plt.xlabel('Company')
plt.ylabel('Value(M)')
plt.title('Bar Chart')
```

```python
plt.bar(col_company,col_Revenue,label='Revenue')
plt.bar(col_company,col_Profit,label='Profit')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.xlabel('Company')
plt.ylabel('Value(M)')
plt.title('Bar Chart')
```

### xticks

```python
# control the x-axis ticks and labels

days = [1,2,3,4,5,6,7]
temperature = [25,30,22,35,24,20,27]

plt.plot(days,temperature)
```

```python
plt.plot(days,temperature)
# labelling xticks
# plt.xticks(xticks,labels)
plt.xticks(days,["One","Two","Three","Four","Five","Six","Seven"])
```

```python
plt.plot(days,temperature)
# change the direction
plt.xticks(days,["One","Two","Three","Four","Five","Six","Seven"],rotation='vertical')
```

```python
plt.plot(days,temperature)
# change the direction
plt.xticks(days,["One","Two","Three","Four","Five","Six","Seven"],rotation=45)
```

## Subplots

### creates multiple subplots within a single figure

```python
plt.subplot()

# plt.subplot(rows,columns,number of the plot)
```

```python
plt.bar(col_company,col_Revenue,label='Revenue')
plt.bar(col_company,col_Profit,label='Profit')
```

```python
# plot 1
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.bar(col_company,col_Revenue,label='Revenue')

# plot 2
plt.subplot(1,2,2)
plt.bar(col_company,col_Profit,label='Profit')
```

```python
fig,ax = plt.subplots(2,2,figsize=(20,10))
# plot our graphs
ax[0,0].bar(col_company,col_Profit,label='Profit')
ax[0,1].bar(col_company,col_Revenue,label='Profit')
ax[1,0].pie(exp_values,labels=exp_labels,autopct='%0.1f%%')
ax[1,1].barh(company,revenue)

plt.suptitle("Matplotlib Graphs")
```
